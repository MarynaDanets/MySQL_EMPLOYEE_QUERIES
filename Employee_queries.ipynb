1. Joining the "employees" and "salaries" tables to retrieve a list of 10 employees, including their emp_no, first_name, last_name, and current salary.

```python
pd.read_sql(f"""
SELECT 
   employees.emp_no, employees.first_name, employees.last_name, salaries.salary FROM employees
join salaries on employees.emp_no = salaries.emp_no
WHERE salaries.to_date = '9999-01-01'
limit 10
""", con)
```

3. Joining the "employees", "dept_emp", and "departments" tables to retrieve a list of 10 employees with emp_no, first_name, last_name, and the department where they currently work.

```python
pd.read_sql(f"""
SELECT 
   employees.emp_no, employees.first_name, employees.last_name, departments.dept_name FROM employees
join dept_emp on employees.emp_no = dept_emp.emp_no
join departments on dept_emp.dept_no = departments.dept_no
WHERE dept_emp.to_date = '9999-01-01'
limit 10
""", con)
```

4. Determining the number of unique employees in each department. Presenting this information in an ordered format, starting with the department that has the highest number of employees.


```python
pd.read_sql(f"""
SELECT
    d.dept_name, COUNT(DISTINCT de.emp_no) as workers FROM departments as d
join dept_emp as de on d.dept_no = de.dept_no
group by d.dept_name
order by workers desc
""", con)
```

<div style="text-align:center; font-size:20px;">Window Functions</div>

1. For the first 10 employees (sorted by `emp_no`), displaying their `emp_no`, `first_name`, `last_name`, and their row_number in the table.

```python
pd.read_sql(f"""
SELECT 
    emp_no, first_name, last_name,
    ROW_NUMBER() OVER (ORDER BY emp_no) as row_num
    FROM employees
limit 10
""", con)
```

2. Displaying the salary progression of the employee with `emp_no = 10001`.

```python
pd.read_sql(f"""
SELECT 
    emp_no, from_date, salary,
    LAG(salary) OVER (ORDER BY from_date) as prev_sal
    FROM salaries
WHERE emp_no = 10001
""", con)
```

3. Determine the salary rank within each department. Displaying the `emp_no`, `dept_name`, current salary, and the employee's rank within their department based on salary.

```python
pd.read_sql(f"""
SELECT s.emp_no, departments.dept_name, s.salary,
    RANK() OVER (PARTITION BY departments.dept_no ORDER BY s.salary desc) as salary_rank
    FROM salaries as s
join dept_emp on s.emp_no = dept_emp.emp_no
join departments on dept_emp.dept_no = departments.dept_no
where s.to_date = '9999-01-01'
limit 10
""", con)
```

4. Find the salary changes for the employee with `emp_no = 10001` over the past 2 years.

```python
pd.read_sql(f"""
SELECT 
    emp_no, salary as current_salary,
    LAG(salary, 2) OVER (ORDER BY from_date) as prev_sal, 
    salary - LAG(salary, 2) OVER (ORDER BY from_date) as difference 
    FROM salaries
WHERE emp_no = 10001
ORDER BY from_date desc
limit 1
""", con)
```

5. Categorize the specified employees into 3 groups based on their current salary. The employees to be analyzed are: `419748, 496734, 264134, 209076, 86631, 456171, 16887, 230255, 246465, 420097, 44362, 280370, 479975, 433320, 473949`.


```python
pd.read_sql(f"""
WITH SelectedEmployees AS (
    SELECT 
        e.emp_no, 
        e.first_name,
        e.last_name,
        s.salary
    FROM employees e
    JOIN salaries s ON e.emp_no = s.emp_no
    WHERE 
        e.emp_no IN (419748, 496734, 264134, 209076, 86631, 456171, 16887, 230255, 246465, 420097, 44362, 280370, 479975, 433320, 473949) AND 
        s.to_date = '9999-01-01'
)

SELECT 
    emp_no, 
    first_name, 
    last_name,
    salary,
    NTILE(3) OVER (ORDER BY salary DESC) as salary_group
FROM SelectedEmployees;
""", con)
```
